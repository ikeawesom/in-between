```python
from Simulation import Simulation
import matplotlib.pyplot as plt
import numpy as np

starting = int(input("Enter buy in: "))
initial = starting
base = initial
x = [initial]
y = [0]

trials = 50
games = 0

for i in range(trials):
    s = Simulation()
    initial = base
    s.setBase(base)
    s.startComputer()
    base = s.getBase()
    if base <= 0:
        break
    profit = base - initial
    wins, loses, passes, win_val, lose_val = s.getStats()

    # print(f"Balance for Game {i + 1} - Start: ${initial} | After: ${base} | Profit: {"+" if profit >= 0 else "-"}${abs(profit)}")
    # print(f"Wins: {wins} (${win_val}) | Loses: {loses} (${lose_val}) | Passes: {passes}\n")
    
    y.append(i + 1)
    x.append(base)
    games += 1

plt.plot(y, x)

plt.xlabel('Games Played')
plt.ylabel('Account Balance ($)')
plt.title('Account Balance for In-Between')
plt.grid(True)

print("="*10, f"After {games} games", "="*10)
print(f"Started with: ${starting}")
print(f"Ended with: ${base}")
print("="*36)

plt.show()
```
